<h5><b>LOAD CÁC MÔ HÌNH</b></h5>

```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
```

<h5><b>ĐẶT BIẾN MỤC TIÊU VÀ CHIA DỮ LIỆU</b></h5>

```python
X = df.drop(columns=['Diabetes_binary'])
y = df['Diabetes_binary']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state=42)
```

<h5><b>XÂY DỰNG GRID SEARCH</b></h5>

```python
# Dictionary scoring với nhiều metric
scoring = {
    'accuracy': 'accuracy',
    'precision': 'precision',
    'recall': 'recall',
    'f1': 'f1'
}
```

```python
# Tập hợp các mô hình và tham số cần đánh giá
models_params = {
    "RandomForest": {
        "model": RandomForestClassifier(random_state=42),
        "params": {
            'max_depth': [3, 5, 7], #Độ sâu tối đa của mỗi cây
            'min_samples_leaf': [1, 3], #Số lượng mẫu tối thiểu tại một lá
            'min_samples_split': [2, 4], #Số lượng mẫu tối thiểu để tiếp tục phân tách tại một nút.
            'max_features': [2, 4], #Số lượng đặc trưng (features) được sử dụng trong mỗi lần tách.
            'n_estimators': [75, 125] #Số lượng cây trong rừng
        }
    },
    "LogisticRegression": {
        "model": LogisticRegression(max_iter=500, random_state=42),
        "params": {
            "C": [0.1, 1, 10], #Hệ số điều chỉnh phạt L2 (regularization)
        }
    },
    "GaussianNB": {
        "model": GaussianNB(),
        "params": {}  # GaussianNB không cần tham số
    },
    "DecisionTree": {
        "model": DecisionTreeClassifier(random_state=42),
        "params": {
            "max_depth": [4, 5, 6, 7, 8, 9, 10, 15, 20, 25],
            "min_samples_split": [2, 5, 10, 20]
        }
    }
}
```

```python
results = []  # Danh sách lưu trữ kết quả từng mô hình
best_models = {}  # Dictionary lưu trữ các mô hình tốt nhất

for name, mp in models_params.items():
    print(f"Đang chạy GridSearchCV cho mô hình: {name}...")
    
    grid_search = GridSearchCV(
        estimator=mp["model"],  # Đối tượng mô hình
        param_grid=mp["params"],  # Tập các giá trị tham số cần tối ưu
        scoring=scoring,
        refit='f1',  # Tối ưu dựa trên f1
        cv=5,  # Áp dụng k-fold cross-validation với k=5 (5 fold)
        n_jobs=-1  # Sử dụng tất cả CPU để tăng tốc quá trình tìm kiếm
    )
    
    grid_search.fit(X_train, y_train)
    
    # Lưu kết quả vào danh sách
    results.append({
        "Model": name,
        "Best Params": grid_search.best_params_,
        "Best Accuracy": grid_search.cv_results_['mean_test_accuracy'].max(),
        "Best Precision": grid_search.cv_results_['mean_test_precision'].max(),
        "Best Recall": grid_search.cv_results_['mean_test_recall'].max(),
        "Best F1": grid_search.cv_results_['mean_test_f1'].max()
    })

    # Lưu lại mô hình tốt nhất vào dictionary
    best_models[name] = grid_search.best_estimator_

# Chuyển kết quả thành DataFrame để so sánh
results_df = pd.DataFrame(results)
print("\nKết quả so sánh các mô hình:")
results_df
```

<b>Dựa vào kết quả trên, ta có thể rút ra được là mô hình random forest là mô hình hiệu quả nhất cho tập dữ liệu</b>

<h5><b>THỰC HIỆN FEATURE SELECTION</b></h5>

```python
# Lấy mô hình RandomForest tốt nhất từ GridSearchCV
best_rf = best_models['RandomForest']

# Truy cập độ quan trọng của các đặc trưng
feature_importances = best_rf.feature_importances_

# Tạo DataFrame để lưu độ quan trọng của các đặc trưng
feature_importance_df = pd.DataFrame({
    'Feature': X_train.columns,  # Các cột (tên đặc trưng) trong tập huấn luyện
    'Importance': feature_importances
}).sort_values(by='Importance', ascending=False)

# Vẽ biểu đồ biểu diễn feature importance
plt.figure(figsize=(10, 6))
plt.barh(feature_importance_df['Feature'], feature_importance_df['Importance'])
plt.gca().invert_yaxis()  # Đảo ngược trục để giá trị lớn nhất ở trên
plt.xlabel("Feature Importance")
plt.ylabel("Features")
plt.title("Feature Importance from RandomForest")
plt.show()
```

<center><h1><b>TỐI ƯU MÔ HÌNH RANDOM FOREST</b></h1></center>

<h5><b>XÂY DỰNG MÔ HÌNH RANDOM FOREST MỚI</b></h5>

```python
X = df.drop(columns=['Diabetes_binary', 'AnyHealthcare', 'NoDocbcCost', 'Fruits', 'Veggies', 'Smoker', 'Stroke', 'Sex', 'CholCheck'])
y = df['Diabetes_binary']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state=42)
```

```python
rf = RandomForestClassifier(random_state = 0)

cv_params = {
        'max_depth': [3,5, None],
        'min_samples_leaf': [1,3],
        'min_samples_split': [2,4],
        'max_features': [2,4],
        'n_estimators': [75,125, ]
}
scoring = ['accuracy', 'precision','recall','f1']

rf_cv = GridSearchCV(rf,
                   cv_params,
                   scoring = ['accuracy', 'precision','recall','f1'],
                   cv = 2,
                   refit = 'f1')

rf_cv.fit(X_train,y_train)
```

```python
rf_cv.best_estimator_
```

<h5><b>ĐÁNH GIÁ MÔ HÌNH RANDOM FOREST</b></h5>

```python
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix 
def conf_matrix_plot(model, x_data, y_data):
    model_pred = model.predict(x_data)
    cm = confusion_matrix(y_data, model_pred, labels=model.classes_)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
    disp.plot(values_format='');
    plt.show()
```

```python
conf_matrix_plot(rf_cv, X_test, y_test)
```

```python
def make_result(model_name, model_object, X_test, y_test):
    # Lấy kết quả từ cross-validation
    cv_results = pd.DataFrame(model_object.cv_results_)
    
    # Tìm thông số tốt nhất dựa trên f1
    best_estimator_results = cv_results.iloc[cv_results['mean_test_f1'].idxmax(), :]
    
    # Dự đoán trên tập kiểm tra
    y_pred = model_object.best_estimator_.predict(X_test)
    
    # Tính các chỉ số
    from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score
    f1 = f1_score(y_test, y_pred)  
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    accuracy = accuracy_score(y_test, y_pred)
    
    # Tạo bảng kết quả
    table = pd.DataFrame({
        'Model': [model_name],
        'f1': [f1],
        'Precision': [precision],
        'Recall': [recall],
        'Accuracy': [accuracy]
    })
    
    return table
```

```python
result_table = make_result("Random Forest", rf_cv, X_test, y_test)
result_table
```
