<center><h1><b>XÂY DỰNG MÔ HÌNH</b></h1></center>

```python
X = df.drop(columns=['Diabetes_binary'])
y = df['Diabetes_binary']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state=42)
```

```python
from sklearn.metrics import make_scorer, precision_score

rf = RandomForestClassifier(random_state = 0)
precision_scorer = make_scorer(precision_score, zero_division=1)

cv_params = {
        'max_depth': [3,5, None],
        'min_samples_leaf': [1,3],
        'min_samples_split': [2,4],
        'max_features': [2,4],
        'n_estimators': [75,125, ]
}

rf_cv = GridSearchCV(
    rf,
    cv_params,
    scoring={
        'accuracy': 'accuracy',
        'precision': precision_scorer,
        'recall': 'recall',
        'f1': 'f1'
    },
    cv=2,
    refit='f1'
)
rf_cv.fit(X_train,y_train)
```

```python
rf_cv.best_estimator_
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix 
def conf_matrix_plot(model, x_data, y_data):
    model_pred = model.predict(x_data)
    cm = confusion_matrix(y_data, model_pred, labels=model.classes_)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
    disp.plot(values_format='');
    plt.show()
```

```python
conf_matrix_plot(rf_cv, X_test, y_test)
```

```python
importances = rf_cv.best_estimator_.feature_importances_
forest_importances = pd.Series(importances, index=X.columns).sort_values(ascending = False)
fig, ax = plt.subplots()
forest_importances.plot.bar(ax=ax)
```

```python
def make_result(model_name, model_object, X_test, y_test):
    # Lấy kết quả từ cross-validation
    cv_results = pd.DataFrame(model_object.cv_results_)
    
    # Tìm thông số tốt nhất dựa trên f1
    best_estimator_results = cv_results.iloc[cv_results['mean_test_f1'].idxmax(), :]
    
    # Dự đoán trên tập kiểm tra
    y_pred = model_object.best_estimator_.predict(X_test)
    
    # Tính các chỉ số
    from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score
    f1 = f1_score(y_test, y_pred)  
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    accuracy = accuracy_score(y_test, y_pred)
    
    # Tạo bảng kết quả
    table = pd.DataFrame({
        'Model': [model_name],
        'f1': [f1],
        'Precision': [precision],
        'Recall': [recall],
        'Accuracy': [accuracy]
    })
    
    return table
```

```python
result_table = make_result("Random Forest", rf_cv, X_test, y_test)
result_table
```
